## since Y1(Heat load) and Y2(Cooling load) are highly correlated seperate models are used for them

# spliting data

```python
from sklearn.linear_model import LinearRegression,PoissonRegressor,Lasso,Ridge,ElasticNet
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor,AdaBoostRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor



from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
max_score=[]
min_score=[]
avg_score=[]
max_score_2=[]
min_score_2=[]
avg_score_2=[]

def record(result):
    min_score.append(np.min(result))
    max_score.append(max(result))
    avg_score.append(np.mean(result))
    print(max_score,min_score,avg_score)
    
def record_2(result):
    min_score_2.append(np.min(result))
    max_score_2.append(max(result))
    avg_score_2.append(np.mean(result))
    print(max_score_2,min_score_2,avg_score_2)
```

```python
model1=LinearRegression()
model2=PoissonRegressor()
model3=Lasso()
model4=Ridge()
model5=ElasticNet()
model6=GradientBoostingRegressor()
model7=RandomForestRegressor()
model8=AdaBoostRegressor()
model9=LGBMRegressor()
model10=CatBoostRegressor()
models=[model1,model2,model3,model4,model5,model6,model7,model8,model9,model10]
```

```python
# models for Heat Loading
for model in models:
    ss=ShuffleSplit(n_splits=20,test_size=0.33,random_state=42)
    result= cross_val_score(model,X,Y['Y1'],cv=ss)
    record(result)
    
```

```python
# models for cooling Loading
for model in models:
    ss=ShuffleSplit(n_splits=20,test_size=0.33,random_state=42)
    result= cross_val_score(model,X,Y['Y2'],cv=ss)
    record_2(result)
```

```python
models=['LinearRegression','PoissonRegressor',
        'Lasso','Ridge','ElasticNet','GradientBoostingRegressor',
        'RandomForestRegressor','AdaBoostRegressor','LGBMRegressor','CatBoostRegressor']
col={'min_score':min_score,'max_score':max_score,"mean_score":avg_score}
col2={'min_score':min_score_2,'max_score':max_score_2,"mean_score":avg_score_2}
df=pd.DataFrame(data=col,index=models)

df2=pd.DataFrame(data=col2,index=models)
```

# Model for heatloading

```python
a=df.sort_values(by=['mean_score'], ascending=False)
a.style.background_gradient(cmap='Blues')
```

```python
ss=ShuffleSplit(n_splits=1,test_size=0.33,random_state=42)
for train_id,test_id in ss.split(data):
    train=data.iloc[train_id]
    test=data.iloc[test_id]
    
```

```python
train.head()
```

```python
x_train =train[['X1','X2','X3','X4','X5','X6','X7','X8']]
y_train =train[['Y1']]
y_train2= train[['Y2']]
```

```python
test.head()
```

```python
x_test =test[['X1','X2','X3','X4','X5','X6','X7','X8']]
y_test =test[['Y1']]
y_test2= test[['Y2']]
```

```python
model =CatBoostRegressor()
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_train)
rmse = (np.sqrt(mean_squared_error(y_train, pred)))
r2 = r2_score(y_train, pred)
print('Training perfomance')
print("RMSE: {}".format(rmse))
print("R2  : {}".format(r2))
```

```python
pred= model.predict(x_test)
rmse = (np.sqrt(mean_squared_error(y_test, pred)))
r2 = r2_score(y_test, pred)
print('Testing perfomance')
print("RMSE: {}".format(rmse))
print("R2  : {}".format(r2))
```

```python
pred= model.predict(X)
rmse = (np.sqrt(mean_squared_error(Y['Y1'], pred)))
r2 = r2_score(Y['Y1'], pred)
print('overall perfomance')
print("RMSE: {}".format(rmse))
print("R2  : {}".format(r2))
```

```python
import pickle
filename="heatload_model.pkl"
pickle.dump(model,open(filename,'wb'))
print("MODEL SAVED SUCESSFULY")
```
